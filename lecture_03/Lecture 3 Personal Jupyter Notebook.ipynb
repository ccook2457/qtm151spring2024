```python
print("How to Generate a Set with random numbers")
print("Loc is the mean, scale is the standard deviation")
randomvar_a = np.random.normal(loc=0, scale=2, size=1000)
print(randomvar_a)
```

```python
np.random.seed(18945)
random_var_b = np.random.normal(loc=0, scale=1, size=10)
print(random_var_b)
```

```python
plt.hist(x = randomvar_a)
plt.xlabel("Variable a")
plt.ylabel("Frequency")
```
